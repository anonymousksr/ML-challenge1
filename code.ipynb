```
#importing the essential python libraries
import pandas as pd
import numpy as np
```

```
#loading the dataset 
df = pd.read_csv(r"kaggledata.csv")
df
```

```
#Getting to know the different columns in the dataset
header = list(df.columns.values)
header
```

```
#Viewing the rows with null values 
df.iloc[10:25,:]
```

```
#Replacing the null values
df.replace('', np.nan, inplace=True)
df
```

```
#Dropping the rows with null values
df.dropna(inplace=True)
df
```

```
#Importing the matplotlib library to plot a graph 
import matplotlib.pyplot as plt 
```

```
#plotting a scatter graph i.e, Attendance versus Total marks of student
plt.scatter( df["Attendance (Marks: 15)"].values , df["Mark total 300"].values , s=5, c='red')
#Naming the X-axis and Y-axis
plt.xlabel('attendance')
plt.ylabel('marks')
#Syntax to view the graph
plt.show()
```

```
#Viewing the datatypes of all columns of the dataframe
df.dtypes
```

```
#To convert all the columns of dataframe to float datatype
df = df.astype(float)
```

```
#To find the correlation factor betwwen Attendance and Total Marks
df["Attendance (Marks: 15)"].corr(df["Mark total 300"])
```

```
#Since the correlation factor is near to zero we can say that there is no relationship between Attendance and the total marks of the student
```
